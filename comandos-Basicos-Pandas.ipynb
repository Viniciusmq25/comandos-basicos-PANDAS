## dropping rows

```python
df.isnull().sum() # mostra a quantidade de valores nulos
```

```python
df.dropna(inplace=True) #exclui as linhas com valores nulos
```

## dropping columns (exclui uma coluna expecifica)

```python
df.drop('Vencedor', axis=1) #exclui a coluna Vencedor
```

```python
df['Total de gols'] = df['Visitante Placar'] + df['Mandante Placar'] #cria uma nova coluna com o total de gols

df
```

```python
df['Coluna toda igual'] = 'Brasileirão Betano' #cria uma nova coluna com o valor igual em todas as linhas

df
```

```python
df.iloc[0,-1] = 'Brasileirão JhonBet' #altera o valor da última coluna da primeira linha

df
```

```python
df['Visitante Venceu'] = df['Vencedor'] == df['Visitante'] #cria uma nova coluna com True se o visitante venceu e False se não
df['Visitante Venceu'].sum() #mostra a quantidade de vezes que o visitante venceu
```

```python
df['Visitante Venceu binario'] = df['Visitante Venceu'].apply(lambda x: 1 if x == True else 0) #cria uma nova coluna com 1 se o visitante venceu e 0 se não

df['Visitante Venceu binario'].mean() #mostra a média de vitórias do visitante
print(f'A média de vitórias do visitante é de {df["Visitante Venceu binario"].mean()*100:.2f}%')
```
